```python
OLP_col
```

```python
# Plot countplots for each column
plt.figure(figsize=(18, 12))
for i, col in enumerate(OLP_col, 1):
    plt.subplot(3, 3, i)
    sns.countplot(data=batch_table, x=col, order = [0,1,2,3,4,5], palette='viridis')
    plt.title(col, fontsize=10)
    plt.xticks(rotation=45)
    plt.tight_layout()
plt.suptitle("Distribution of Responses - Organization Leadership & Planning", fontsize=16, y=1.02)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(20, 14))

for i, col in enumerate(OLP_col, 1):
    plt.subplot(3, 3, i)
    sns.boxplot(data=batch_table, x='What is your age?', y=col, palette='coolwarm')
    plt.title(f"Boxplot: {col} vs Age Group", fontsize=10)
    plt.xticks(rotation=45)

plt.tight_layout()
plt.suptitle("Boxplots for Organisation Leadership & Planning Responses Across Age Groups", fontsize=16, y=1.02)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(batch_table[OLP_col].corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Between Organisation Leadership & Planning Questions")
plt.show()
```

```python
age_summary = batch_table.groupby('What is your age?')[OLP_col].mean()
age_summary.plot(kind='bar', figsize=(16, 8), colormap='viridis')
plt.title("Average Response per Organisation Leadership & Planning Question by Age Group")
plt.ylabel("Average Score (0 = Neutral, 5 = Strongly Agree)")
plt.xticks(rotation=45)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Radar plot setup
age_means = batch_table.groupby('What is your age?')[OLP_col].mean()
labels = OLP_col
num_vars = len(labels)

angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
labels += [labels[0]]
angles += [angles[0]]

plt.figure(figsize=(15, 15))

for index, row in age_means.iterrows():
    values = row.tolist()
    values += [values[0]]  # close the loop
    plt.polar(angles, values, label=index, linewidth=2)

plt.xticks(angles[:-1], labels, fontsize=10)
plt.title("Radar Chart: OLP Responses by Age Group", size=14)
plt.legend(loc='upper right', bbox_to_anchor=(1.3, 1.1))
plt.tight_layout()
plt.show()
```
